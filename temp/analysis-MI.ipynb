```python
import matplotlib.pyplot as plt
import numpy as np

# Set up the figure and axis
fig, ax = plt.subplots()

# Draw the square
square_x = [0, 100, 100, 0, 0]
square_y = [0, 0, 100, 100, 0]
ax.plot(square_x, square_y, label='Square')

# Draw the inscribed hexagon
hexagon_radius = 50  # Half of the side length of the square
hexagon_angle = np.linspace(0, 2 * np.pi, 7)
hexagon_x = 50 + hexagon_radius * np.cos(hexagon_angle)
hexagon_y = 50 + hexagon_radius * np.sin(hexagon_angle)
ax.plot(hexagon_x, hexagon_y, label='Inscribed Hexagon')

# Draw the inscribed circle of the hexagon
circle_radius = hexagon_radius * np.cos(np.pi / 6)  # Radius of the circle
circle = plt.Circle((50, 50), circle_radius, fill=False, edgecolor='r', label='Inscribed Circle')
ax.add_patch(circle)

# Draw the square inscribed in the circle
inscribed_square_side = circle_radius * np.sqrt(2)
inscribed_square_half_side = inscribed_square_side / 2
inscribed_square_x = [50 - inscribed_square_half_side, 50 + inscribed_square_half_side,
                      50 + inscribed_square_half_side, 50 - inscribed_square_half_side,
                      50 - inscribed_square_half_side]
inscribed_square_y = [50 - inscribed_square_half_side, 50 - inscribed_square_half_side,
                      50 + inscribed_square_half_side, 50 + inscribed_square_half_side,
                      50 - inscribed_square_half_side]
ax.plot(inscribed_square_x, inscribed_square_y, label='Inscribed Square in Circle')

# Set axis limits and labels
ax.set_xlim(-10, 110)
ax.set_ylim(-10, 110)
ax.set_aspect('equal', 'box')
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)
ax.grid(color='gray', linestyle='--', linewidth=0.5)

# Add legend
ax.legend(loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol=2)

# Show plot
plt.show()
```
